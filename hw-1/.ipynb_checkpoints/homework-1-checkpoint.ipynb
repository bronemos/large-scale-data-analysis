### Problem Setup

```python
import time
import numpy as np
import matplotlib.pyplot as plt

dim = 100
n_iter = 5000

np.random.seed(337)
# Generating positive semidefinite matrix
A = np.random.rand(dim, dim)
A = A @ A.T

b = np.random.rand(dim, 1)
lr = 0

# Calculating optimum directly using matrix inverse
x_opt = np.linalg.inv(A) @ b

L = np.max(np.linalg.eigvals(A))
mu = np.min(np.linalg.eigvals(A))
alpha = 1/L
chi = L/mu
```

```python
# Function that we are optimizing
def function(x, A=A, b=b):
    return 1/2 * x.T @ A @ x - b.T @ x
```

```python
# Function for plotting descents
def plot_desc(x_values, n_iter, title, x_opt=x_opt):
    conv = np.log(np.linalg.norm(x_values.T - x_opt, ord=2, axis=0))
    plt.figure(figsize=(20,10))
    plt.xlabel("Iteration")
    plt.ylabel("log||x_k - x*||")
    plt.title(title)
    plt.plot(range(n_iter), conv)
    plt.savefig(title.lower().replace(" ", "_") + ".png")
    plt.show()
```

### Gradient Descent Algorithm

```python
def gradient_descent(alpha, n_iter=5000, tolerance=1e-5):
    x = np.zeros((dim, 1))
    x_values = np.empty((n_iter, dim))
    for iter in range(n_iter):
        iters = iter
        grad = A @ x - b
        x -= alpha * grad
        x_values[iter, :] = x.flatten()
        if np.linalg.norm(grad, ord=2) <= tolerance:
            break
    return x_values[:iters + 1], iters + 1

%time x_steep, iters = gradient_descent(alpha)
plot_desc(x_steep, iters, "Gradient Descent")
```

### Conjugate Gradient Algorithm

```python
def conjugate_gradient(alpha, n_iter=5000, tolerance=1e-5):
    x = np.zeros((dim, 1))
    x_values = np.empty((n_iter, dim))
    for iter in range(n_iter):
        iters = iter
        if iter == 0:
            grad = A @ x - b
            rho = -grad
            x += alpha * rho
        else:
            delta = np.linalg.norm(A @ x - b, ord=2) / np.linalg.norm(grad, ord=2)
            grad = A @ x - b
            rho = -grad + delta * rho
            if function(x) <= function(x + alpha * rho):
                rho = -grad
            x += alpha * rho
        x_values[iter, :] = x.flatten()
        if np.linalg.norm(grad, ord=2) <= tolerance:
            break
    return x_values[:iters + 1], iters + 1

%time x_conj, iters = conjugate_gradient(alpha)
plot_desc(x_conj, iters, "Conjugate Gradient")
```

### Nesterov's I Algorithm

```python
def nesterovs(alpha, n_iter=5000, tolerance=1e-5):
    x_curr = np.zeros((dim, 1))
    y = x_curr
    x_values = np.empty((n_iter, dim))
    beta = alpha * (1 - alpha) / (alpha ** 2 + alpha)
    for iter in range(n_iter):
        iters = iter
        grad = A @ x_curr - b
        grad_y = A @ y - b
        x_prev = x_curr
        x_curr = y - alpha * grad_y
        y = x_curr + beta * (x_curr - x_prev)
        x_values[iter, :] = x_curr.flatten()
        if np.linalg.norm(grad, ord=2) <= tolerance:
            break
    return x_values[:iters + 1], iters + 1
    
%time x_nesterovs, iters = nesterovs(alpha)
plot_desc(x_nesterovs, iters, "Nesterov's I")
```

### Coordinate Descent Algorithm

```python
def coord_desc(alpha, stochastic=False, n_iter=5000, tolerance=1e-5):
    x = np.zeros((dim, 1))
    x_values = np.empty((n_iter, dim))
    for iter in range(n_iter):
        iters = iter
        if stochastic:
            i = np.random.randint(dim)
        else:
            i = iter % dim
        grad = A[i, :] @ x - b[i]
        x[i] -= alpha * grad
        x_values[iter, :] = x.flatten()
        if np.linalg.norm(grad, ord=2) <= tolerance:
            break
    return x_values[:iters + 1], iters + 1

%time x_coord, iters = coord_desc(alpha)
plot_desc(x_coord, iters, "Coordinate Descent - Deterministic")

%time x_coord, iters = coord_desc(alpha, stochastic=True)
plot_desc(x_coord, iters, "Coordinate Descent - Stochastic")
```
